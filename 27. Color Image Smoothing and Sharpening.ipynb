```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def smooth_and_sharpen(image_path):
    # Load the image
    img = cv2.imread(image_path)

    # Convert BGR image to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Smoothing techniques
    # Gaussian blur
    gaussian_blurred = cv2.GaussianBlur(img_rgb, (9, 9), 0)
    
    # Median blur
    median_blurred = cv2.medianBlur(img_rgb, 5)
    
    # Bilateral filter
    bilateral_filtered = cv2.bilateralFilter(img_rgb, 9, 75, 75)

    # Sharpening techniques
    # Laplacian kernel for sharpening
    laplacian_kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
    laplacian_sharpened = cv2.filter2D(img_rgb, -1, laplacian_kernel)

    # Unsharp masking
    gaussian_blur_for_unsharp = cv2.GaussianBlur(img_rgb, (9, 9), 10)
    unsharp_masked = cv2.addWeighted(img_rgb, 1.5, gaussian_blur_for_unsharp, -0.5, 0)

    # Visualization
    plt.figure(figsize=(20, 12))
    
    plt.subplot(231), plt.imshow(img_rgb), plt.title('Original Image')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(232), plt.imshow(gaussian_blurred), plt.title('Gaussian Blur')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(233), plt.imshow(median_blurred), plt.title('Median Blur')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(234), plt.imshow(bilateral_filtered), plt.title('Bilateral Filter')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(235), plt.imshow(laplacian_sharpened), plt.title('Laplacian Sharpened')
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(236), plt.imshow(unsharp_masked), plt.title('Unsharp Masking')
    plt.xticks([]), plt.yticks([])

    plt.show()

# Path to your image
image_path = 'c130.jpg'
smooth_and_sharpen(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('c130.jpg')

# Convert to HSV for some techniques
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# -------- Smoothing Techniques --------
# Gaussian Blur
img_gaussian = cv2.GaussianBlur(img, (5, 5), 0)

# Median Blur
img_median = cv2.medianBlur(img, 5)

# Bilateral Filter
img_bilateral = cv2.bilateralFilter(img_hsv, 9, 75, 75)

# -------- Sharpening Techniques --------
# Unsharp masking 
gaussian_3 = cv2.GaussianBlur(img, (9, 9), 10.0)  # Blurred version
unsharp_image = cv2.addWeighted(img, 1.5, gaussian_3, -0.5, 0)

# Laplacian Filter
kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
img_laplacian = cv2.filter2D(img, -1, kernel)

# -------- Visualization --------
images = [img, img_gaussian, img_median, img_bilateral, unsharp_image, img_laplacian]
titles = ['Original', 'Gaussian Blur', 'Median Blur', 'Bilateral Filter', 'Unsharp Masking', 'Laplacian']

for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
