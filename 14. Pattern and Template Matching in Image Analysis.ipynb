```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalized_cross_correlation(img, template):
    """Calculates the NCC between an image and a template"""

    img_mean = np.mean(img)
    template_mean = np.mean(template)

    img_std = np.std(img)
    template_std = np.std(template)

    # Calculate the zero-mean versions of the image and template
    img_zm = img - img_mean
    template_zm = template - template_mean

    # Calculate the cross-correlation (numerator)
    numerator = np.sum((img_zm) * (template_zm))

    # Calculate the denominator
    denominator = img_std * template_std * img.size

    ncc = numerator / denominator
    return ncc

# Load the image and template
img = cv2.imread('c130.jpg', cv2.IMREAD_GRAYSCALE)
template = cv2.imread('c130_modified.jpg', cv2.IMREAD_GRAYSCALE)  # Replace 'template.jpg' with your template image
template = cv2.resize(template, (0, 0), fx=0.5, fy=0.5)  # Resize by 50%

# Calculate the NCC map
result = cv2.matchTemplate(img, template, cv2.TM_CCOEFF_NORMED)

# Find the location of the best match
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
top_left = max_loc
bottom_right = (top_left[0] + template.shape[1], top_left[1] + template.shape[0])

# Draw a rectangle around the detected region
cv2.rectangle(img, top_left, bottom_right, color=(0, 255, 0), thickness=2)

# Visualization
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Detected Template'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, cmap='gray')
plt.title('NCC Matching Result'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def ssd_matching(img, template):
    """Calculates the SSD matching result between an image and a template"""

    # Initialize a 2D array for storing SSD scores with the correct shape
    result = np.zeros((img.shape[0] - template.shape[0] + 1,
                       img.shape[1] - template.shape[1] + 1))

    print("Image Shape:", img.shape)                 # For debugging
    print("Template Shape:", template.shape)         # For debugging

    # Iterate over each (i, j) position where the template can fit within the image
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            # Extract the current patch of the image to compare with the template
            image_patch = img[i:i + template.shape[0], j:j + template.shape[1]]
            # Calculate the SSD for the current patch and template
            ssd = np.sum((image_patch - template) ** 2)
            # Store the SSD result in the 2D result array
            result[i, j] = ssd

    print("Result Array Shape:", result.shape)      # For debugging
    return result





# Load the image and template
img = cv2.imread('c130.jpg', cv2.IMREAD_GRAYSCALE)
template = cv2.imread('c130_modified.jpg', cv2.IMREAD_GRAYSCALE)  


# Resize the template if necessary
template = cv2.resize(template, (0, 0), fx=0.5, fy=0.5)  # Reduce size by 50%

# Continue with the rest of the code ...

# Calculate the SSD map
result = ssd_matching(img, template)

# Find the location of the best match (lowest SSD)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
top_left = min_loc
bottom_right = (top_left[0] + template.shape[1], top_left[1] + template.shape[0])

# Draw a rectangle around the detected region
cv2.rectangle(img, top_left, bottom_right, color=(0, 255, 0), thickness=2)

# Visualization - Reshaping integrated here
result_image = result.reshape(img.shape[0] - template.shape[0] + 1,
                              img.shape[1] - template.shape[1] + 1) 

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Detected Template'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result_image, cmap='gray') # Use the reshaped array
plt.title('SSD Matching Result'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def zncc_matching(img, template):
   """Calculates the ZNCC matching result between an image and a template"""

   # Calculate means of the image and template
   img_mean = np.mean(img)
   template_mean = np.mean(template)

   # Subtract means to create zero-mean versions
   img_zm = img - img_mean
   template_zm = template - template_mean

   # Calculate standard deviations
   img_std = np.std(img_zm)
   template_std = np.std(template_zm)

   # Initialize a 2D array for storing ZNCC scores with the correct shape
   result = np.zeros((img.shape[0] - template.shape[0] + 1,
                     img.shape[1] - template.shape[1] + 1))

   # Iterate over each (i, j) position where the template can fit within the image
   for i in range(result.shape[0]):
      for j in range(result.shape[1]):
         # Extract the current patch of the image to compare with the template
         image_patch_zm = img_zm[i:i + template.shape[0], j:j + template.shape[1]]

         # Calculate the ZNCC for the current patch and template
         zncc = np.sum(image_patch_zm * template_zm) / (img_std * template_std)

         # Store the ZNCC result in the 2D result array
         result[i, j] = zncc

   return result

# Load the image and template
img = cv2.imread('c130.jpg', cv2.IMREAD_GRAYSCALE)
template = cv2.imread('c130_modified.jpg', cv2.IMREAD_GRAYSCALE)

# Resize the template if necessary
template = cv2.resize(template, (0, 0), fx=0.5, fy=0.5)

# Calculate the ZNCC map
result_image = zncc_matching(img, template)
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Detected Template'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result_image, cmap='gray') # Use the reshaped array
plt.title('SSD Matching Result'), plt.xticks([]), plt.yticks([])
plt.show()
```
