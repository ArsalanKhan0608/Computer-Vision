```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt 
image_path = "c130.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # OpenCV loads as BGR

# Display the original image
plt.figure(figsize=(8, 5))
plt.imshow(image)
plt.title("Original Image")
plt.show()
def gray_world(image):
    avg_color = np.mean(image, axis=(0,1)) 
    avg_gray = np.mean(avg_color)
    scale_factors = avg_gray / avg_color  
    return np.clip(image * scale_factors[None, None, :], 0, 255).astype('uint8')
def white_patch(image):
    max_vals = np.max(image, axis=(0, 1))  
    scale_factors = 255.0 / max_vals   
    return np.clip(image * scale_factors[None, None, :], 0, 255).astype('uint8')
def shades_of_gray(image, p=6):  
    img_log = np.log(image + 1)
    minkowski_norm = np.power(np.sum(np.power(img_log, p), axis=2), 1.0/p) 
    gray_avg = np.mean(minkowski_norm)
    scale_factors = gray_avg / minkowski_norm[:,:,None]  
    return np.clip(image * scale_factors, 0, 255).astype('uint8')
def complex_color_constancy(image, w1=0.3, w2=0.3, w3=0.4):
    """
    Applies a weighted combination of Gray World, White Patch, and Shades of Gray 
    """
    corrected_gw = gray_world(image)
    corrected_wp = white_patch(image)
    corrected_sg = shades_of_gray(image)

    result = (w1 * corrected_gw + w2 * corrected_wp + w3 * corrected_sg).astype('uint8')
    return result
corrected_gw = gray_world(image.copy())
corrected_wp = white_patch(image.copy())
corrected_sg = shades_of_gray(image.copy())
corrected_complex = complex_color_constancy(image.copy())

# Visualization
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.imshow(corrected_gw)
plt.title("Gray World")

plt.subplot(2, 2, 2)
plt.imshow(corrected_wp)
plt.title("White Patch")

plt.subplot(2, 2, 3)
plt.imshow(corrected_sg)
plt.title("Shades of Gray")

plt.subplot(2, 2, 4)
plt.imshow(corrected_complex)
plt.title("Complex Combination")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gray_world_algorithm(image):
    """Applies the Gray World color constancy algorithm on the image."""
    avg_b = np.mean(image[:, :, 0])
    avg_g = np.mean(image[:, :, 1])
    avg_r = np.mean(image[:, :, 2])

    avg_gray = (avg_b + avg_g + avg_r) / 3

    image[:, :, 0] = np.clip(image[:, :, 0] * (avg_gray / avg_b), 0, 255)
    image[:, :, 1] = np.clip(image[:, :, 1] * (avg_gray / avg_g), 0, 255)
    image[:, :, 2] = np.clip(image[:, :, 2] * (avg_gray / avg_r), 0, 255)
    return image

def white_patch_algorithm(image):
    """Applies the White Patch color constancy algorithm on the image."""
    max_b = np.max(image[:, :, 0])
    max_g = np.max(image[:, :, 1])
    max_r = np.max(image[:, :, 2])

    image[:, :, 0] = np.clip(image[:, :, 0] * (255 / max_b), 0, 255)
    image[:, :, 1] = np.clip(image[:, :, 1] * (255 / max_g), 0, 255)
    image[:, :, 2] = np.clip(image[:, :, 2] * (255 / max_r), 0, 255)
    return image

def retinex_algorithm(image):
    """Applies a simple version of the Retinex color constancy algorithm on the image."""
    image_log = np.log1p(image)
    for i in range(3):
        image_log[:, :, i] = image_log[:, :, i] - np.mean(image_log[:, :, i])

    image_retinex = np.clip(np.exp(image_log) - 1, 0, 255).astype(np.uint8)
    return image_retinex

def visualize_images(original_image, modified_images, titles):
    """Visualizes the original and modified images."""
    plt.figure(figsize=(20, 10))
    plt.subplot(1, len(modified_images) + 1, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')

    for i, (image, title) in enumerate(zip(modified_images, titles), start=2):
        plt.subplot(1, len(modified_images) + 1, i)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title(title)

    plt.show()

# Load the image
image_path = 'c130.jpg'
original_image = cv2.imread(image_path)

# Apply color constancy algorithms
gray_world_img = gray_world_algorithm(original_image.copy())
white_patch_img = white_patch_algorithm(original_image.copy())
retinex_img = retinex_algorithm(original_image.copy())

# Visualize the original and modified images
modified_images = [gray_world_img, white_patch_img, retinex_img]
titles = ['Gray World', 'White Patch', 'Retinex']
visualize_images(original_image, modified_images, titles)
```
