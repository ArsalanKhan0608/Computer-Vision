```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Helper Functions
# --------------------------

def display_images(images, titles, cmap=None):
    """Displays multiple images with titles"""
    assert len(images) == len(titles)
    rows = 1
    cols = len(images)
    for i, (image, title) in enumerate(zip(images, titles)):
        plt.subplot(rows, cols, i+1)
        plt.imshow(image, cmap=cmap)
        plt.title(title)
        plt.xticks([]), plt.yticks([])
    plt.show()

# --------------------------
# Load Image
# --------------------------

img_path = 'c130.jpg'  # Replace with your image path
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Ensure RGB format

# --------------------------
# Color Balancing Techniques
# --------------------------

# 1. Gray World Assumption
def gray_world(img):
    avg_color_per_row = np.average(img, axis=0)
    avg_color = np.average(avg_color_per_row, axis=0)
    scale_factors = avg_color / avg_color.mean()
    img_balanced = (img * scale_factors[None, None, :]).astype(np.uint8)
    return img_balanced

# 2. White Patch / Max White
def white_patch(img):
    max_per_channel = np.max(img, axis=(0, 1))
    scale_factors = 255.0 / max_per_channel
    img_balanced = (img * scale_factors[None, None, :]).astype(np.uint8)
    return img_balanced

# 3. Histogram Stretching
def histogram_stretching(img):
    img_yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])
    img_balanced = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
    return img_balanced

# 4. Singular Value Decomposition (SVD)
def svd_balancing(img):
    r, g, b = cv2.split(img)
    for channel in (r, g, b):
        m, s = np.mean(channel), np.std(channel) 
        channel = (channel - m) / s * 100 + m  # Simulate standard deviation adjustment
    img_balanced = cv2.merge([r, g, b])
    return img_balanced.astype(np.uint8)

# --------------------------
# Apply and Visualize
# --------------------------

img_gray_world = gray_world(img.copy())
img_white_patch = white_patch(img.copy())
img_hist_stretched = histogram_stretching(img.copy())
img_svd = svd_balancing(img.copy())

display_images(
    [img, img_gray_world, img_white_patch, img_hist_stretched, img_svd],
    ['Original', 'Gray World', 'White Patch', 'Hist. Stretching', 'SVD']
)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure

# Load the image
image_path = 'c130.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Function to display images
def show_images(images, titles):
    plt.figure(figsize=(15, 5))
    for i, (img, title) in enumerate(zip(images, titles), 1):
        plt.subplot(1, len(images), i)
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')
    plt.show()

# 1. Simple color balance
def simple_color_balance(img, percent=1):
    assert percent > 0 and percent < 100, 'Percent must be between 0 and 100'
    out_channels = []
    for channel in cv2.split(img):
        flat = channel.flatten()
        low, high = np.percentile(flat, percent), np.percentile(flat, 100-percent)
        channel = np.clip(channel, low, high)
        channel = cv2.normalize(channel, None, 0, 255, cv2.NORM_MINMAX)
        out_channels.append(channel)
    return cv2.merge(out_channels)

# 2. Gray world assumption
def gray_world_assumption(img):
    avg_b, avg_g, avg_r = cv2.mean(img)[:3]
    gray_value = (avg_b + avg_g + avg_r) / 3
    scale = gray_value / np.array([avg_b, avg_g, avg_r])
    out_channels = [img[:, :, i] * scale[i] for i in range(3)]
    out_img = cv2.merge(out_channels)
    return np.clip(out_img, 0, 255).astype(np.uint8)

# 3. Histogram equalization
def histogram_equalization(img):
    img_yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])
    hist_eq_img = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
    return hist_eq_img

# 4. Contrast Limited Adaptive Histogram Equalization (CLAHE)
def clahe(img):
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img_lab = cv2.cvtColor(img, cv2.COLOR_RGB2Lab)
    img_lab[:, :, 0] = clahe.apply(img_lab[:, :, 0])
    clahe_img = cv2.cvtColor(img_lab, cv2.COLOR_Lab2RGB)
    return clahe_img

# 5. Retinex algorithm - Using built-in library for complex implementation
def retinex_with_adjust(img):
    img_float = img.astype(np.float32) / 255.0
    retinex = exposure.adjust_log(img_float, 1)
    return (retinex * 255).astype(np.uint8)

# Applying color balancing techniques
simple_balanced = simple_color_balance(image)
gray_world_balanced = gray_world_assumption(image)
hist_eq = histogram_equalization(image)
clahe_balanced = clahe(image)
retinex_balanced = retinex_with_adjust(image)

# Visualizing the original and modified images
show_images([image, simple_balanced, gray_world_balanced, hist_eq, clahe_balanced, retinex_balanced],
            ['Original', 'Simple Balance', 'Gray World', 'Histogram Equalization', 'CLAHE', 'Retinex'])
```
