```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage.segmentation import slic, felzenszwalb, mark_boundaries 
from skimage.future import graph 
def load_image(image_path):
    img = mpimg.imread(image_path)

    # (Optional) Convert to grayscale if needed
    if len(img.shape) == 3:
        img = np.mean(img, axis=2) 

    return img

img = load_image('c130.jpg')
plt.figure(figsize=(8, 6))
plt.imshow(img, cmap='gray') 
plt.title('Original Image')
plt.show()
num_segments = 200  # You can adjust the number of superpixels
segments = felzenszwalb(img, scale=100, sigma=0.8, min_size=50)  

# Visualize initial segmentation
plt.figure(figsize=(8, 6))
plt.imshow(mark_boundaries(img, segments))
plt.title('Initial Superpixel Segmentation')
plt.show() 
def compute_superpixel_features(img, segments):
    # Placeholder - Example Features 
    features = np.zeros((segments.max() + 1, 2))  
    for i in range(segments.max() + 1):
        features[i, 0] = np.mean(img[segments == i])  # Mean intensity
        # ... You'd add more features (texture, etc.) ...

    return features

features = compute_superpixel_features(img, segments)
def pairwise_potential(features, i, j):
    # Example: Encourage smoothness based on intensity similarity
    return np.exp(-abs(features[i, 0] - features[j, 0]) / 0.2)  

# (Unary potentials would require modeling or learning and are omitted for brevity)
# Requires installation: pip install pygco
import pygco 

edges = graph.rag_mean_color(img, segments)  # Create edges between superpixels
labels = pygco.cut_from_graph(edges, features, pairwise_potential) 
segmented_img = np.choose(labels, segments)  

plt.figure(figsize=(8, 6))
plt.imshow(segmented_img)
plt.title('Segmented Image')
plt.show() 
```

```python
import numpy as np
import cv2
import maxflow
import matplotlib.pyplot as plt

def segment_image_using_mrf(image_path):
    # Load the image
    img = cv2.imread(image_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Set up the graph
    g = maxflow.Graph[float]()
    nodeids = g.add_grid_nodes(img_gray.shape)

    # Add non-terminal edges with the same capacity.
    g.add_grid_edges(nodeids, 50)

    # Add the terminal edges. The image intensity is used to compute capacities.
    g.add_grid_tedges(nodeids, img_gray, 255-img_gray)

    # Find the maximum flow/min-cut.
    g.maxflow()
    sgm = g.get_grid_segments(nodeids)

    # The output image
    img_segmented = np.int_(np.logical_not(sgm))

    # Visualize the original and segmented images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Segmented Image')
    plt.imshow(img_segmented, cmap='gray')
    plt.axis('off')

    plt.show()

# Path to the image
image_path = 'c130.jpg'

# Perform segmentation
segment_image_using_mrf(image_path)
```
