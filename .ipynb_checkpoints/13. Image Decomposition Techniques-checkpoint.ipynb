```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import pywt

# Load the image
image_path = 'c130.jpg'
image = plt.imread(image_path)

# Convert to grayscale if it's a color image
if len(image.shape) > 2:
    image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])

# Display the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Apply Fourier Transform
image_fft = fft.fft2(image)
image_fft_shift = fft.fftshift(image_fft)
image_fft_abs = np.abs(image_fft_shift)

# Display the Fourier Transform
plt.subplot(1, 3, 2)
plt.imshow(np.log(image_fft_abs), cmap='gray', extent=(-np.pi, np.pi, -np.pi, np.pi))
plt.title('Fourier Transform')

# Apply Wavelet Transform
coeffs = pywt.dwt2(image, 'haar')
cA, (cH, cV, cD) = coeffs

# Display the Wavelet Transform
plt.subplot(1, 3, 3)
plt.imshow(cA, cmap='gray')
plt.title('Wavelet Transform')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA
from scipy.misc import face

# Load the image
image = face(gray=True)

# Flatten the image
X = image.reshape((-1, image.shape[2] if len(image.shape) > 2 else 1))


# Apply PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

# Apply ICA
ica = FastICA(n_components=1)
X_ica = ica.fit_transform(X)

# Display the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Display the PCA result
plt.subplot(1, 3, 2)
plt.imshow(X_pca.reshape(image.shape), cmap='gray')
plt.title('PCA')

# Display the ICA result
plt.subplot(1, 3, 3)
plt.imshow(X_ica.reshape(image.shape), cmap='gray')
plt.title('ICA')

plt.show()
```

```python
import numpy as np
from scipy import signal, fft
from PIL import Image
import matplotlib.pyplot as plt

# --- Image Loading ---
def load_image(path):
    img = Image.open(path).convert('L')  # Convert to grayscale
    return np.array(img)

# --- Visualization Utility ---
def show_images(images, titles=None):
    num_images = len(images)
    if titles is None:
        titles = [f"Image {i+1}" for i in range(num_images)]
    
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, num_images, i+1)
        plt.imshow(img, cmap='gray')
        plt.title(title)   
    plt.show()

# --- Image & Paths ---
image_path = 'c130.jpg'
image = load_image(image_path)

# --- Fourier Analysis ---
fourier_transform = fft.fft2(image)
fourier_magnitude = np.abs(fourier_transform)

# --- Short-Time Fourier Transform (STFT) ---
window_size = 64  # Adjust as needed
window = signal.windows.hann(window_size)
stft = signal.stft(image, window=window, nperseg=window_size)
frequencies, times, stft_result = stft

stft_magnitude = np.abs(stft_result)

# --- Discrete Wavelet Transform (DWT) ---
coeffs = pywt.wavedec2(image, 'haar', level=2)  # Example decomposition
approx = pywt.waverec2(coeffs[:-1], 'haar')  # Remove detail coefficients


# --- Visualization ---

# Select a time window for STFT magnitude  
stft_magnitude_at_time0 = stft_magnitude[:, 0, :] 

show_images([image, np.log(fourier_magnitude), stft_magnitude_at_time0, approx], 
            ['Original Image', 'Fourier Magnitude (Log)', 'STFT Magnitude (Time 0)', 'DWT Approximation'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, img_as_ubyte
from skimage.restoration import denoise_tv_chambolle
import pywt

# Load the image
image_path = 'c130.jpg'
image = io.imread(image_path)

# Convert to grayscale if it's a color image
if len(image.shape) > 2:
    image = color.rgb2gray(image)

# Apply MCA (Wavelet Transform)
coeffs = pywt.dwt2(image, 'haar')
cA, (cH, cV, cD) = coeffs

# Apply Variational Decomposition (Total Variation Denoising)
image_vd = denoise_tv_chambolle(image, weight=0.1)

# Display the original image
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Display the MCA result
plt.subplot(1, 3, 2)
plt.imshow(cA, cmap='gray')
plt.title('MCA')

# Display the Variational Decomposition result
plt.subplot(1, 3, 3)
plt.imshow(image_vd, cmap='gray')
plt.title('Variational Decomposition')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.decomposition import MiniBatchDictionaryLearning
from skimage.restoration import denoise_tv_chambolle

# Load and preprocess the image
image_path = 'c130.jpg'
image = Image.open(image_path).convert('L')
image = np.array(image)

# Perform dictionary learning (a type of MCA)
dico = MiniBatchDictionaryLearning(n_components=2, alpha=1, max_iter=500)
V = dico.fit_transform(image.reshape(-1, 1))
components = dico.components_

# Perform total variation denoising (a type of variational method)
vd_image = denoise_tv_chambolle(image, weight=0.1)

# Display the results
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')
axes[1].imshow(components[0].reshape(image.shape), cmap='gray')
axes[1].set_title('Component 1')
axes[2].imshow(components[1].reshape(image.shape), cmap='gray')
axes[2].set_title('Component 2')
axes[3].imshow(vd_image, cmap='gray')
axes[3].set_title('Variational Decomposition')
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize
from skimage import io, img_as_float
from skimage.transform import radon, iradon  
import matplotlib.pyplot as plt
from skimage.restoration import denoise_wavelet

patch_size = 8  # Choose your desired patch size

def learn_dictionary(image, n_atoms=256, n_iter=100):
    """Implements a simplified dictionary learning algorithm (can be improved 
    with K-SVD or other methods)"""

    patches = extract_image_patches(image)
    patches = patches - np.mean(patches, axis=1, keepdims=True) 

    dictionary = np.random.randn(patches.shape[1], n_atoms)

    for _ in range(n_iter):
        codes = l0_thresholding(np.dot(dictionary.T, patches), sparsity=0.1)
        dictionary = np.dot(patches, codes.T) 
        dictionary /= np.linalg.norm(dictionary, axis=0, keepdims=True) 

    return dictionary

def extract_image_patches(image, patch_size=8):
    patches = image.reshape(image.shape[0] // patch_size, patch_size,
                            image.shape[1] // patch_size, patch_size)
    patches = patches.transpose(0, 2, 1, 3)
    return patches.reshape(-1, patch_size**2)

def l0_thresholding(coefficients, sparsity):
    """Approximate L0-norm sparse coding - sets all coefficients 
    below a threshold to zero."""
    threshold = np.sort(np.abs(coefficients))[::-1][int(sparsity * coefficients.size)]
    coefficients[np.abs(coefficients) < threshold] = 0
    return coefficients

def mca(image, dictionaries, alpha):
    """Performs Morphological Component Analysis"""
    components = []

    for dictionary in dictionaries:
        patches = extract_image_patches(image)
        patches = patches - np.mean(patches, axis=1, keepdims=True)

        def objective(x):
            representation = np.dot(dictionary, x.reshape(-1, 1)).reshape(-1)
            reconstruction_error = 0.5 * np.linalg.norm(patches - representation.reshape(-1, 1))**2
            l1_regularization = alpha * np.linalg.norm(x, ord=1)
            return reconstruction_error + l1_regularization

        result = minimize(objective, np.zeros(dictionary.shape[1]), method='L-BFGS-B')
        codes = result.x

        component = np.zeros_like(image)
        for i, patch in enumerate(patches):
            component_patch = np.dot(dictionary, codes[i]).reshape(patch.shape)
            row = (i // (image.shape[1] // patch_size)) * patch.shape[0]
            col = (i % (image.shape[1] // patch_size)) * patch.shape[1]
            component[row:row + patch.shape[0], col:col + patch.shape[1]] = component_patch

        components.append(component)

    return components

# -------------- Load Example Image -------------- 
image_path = "c130.jpg"  
image = img_as_float(io.imread(image_path, as_gray=True))

def pad_image(image, patch_size):
    pad_h = patch_size - (image.shape[0] % patch_size)
    pad_w = patch_size - (image.shape[1] % patch_size)
    pad_dims = ((0, pad_h), (0, pad_w))  
    return np.pad(image, pad_dims, mode='constant')

# Load and potentially pad before proceeding
image = img_as_float(io.imread(image_path, as_gray=True))
if image.shape[0] % patch_size != 0 or image.shape[1] % patch_size != 0:
    image = pad_image(image, patch_size)

# -------------- Learn Dictionaries -------------- 
# Let's create one dictionary using wavelets and one using a Radon transform  
wavelet_dictionary = denoise_wavelet(image, method='BayesShrink', mode='soft', multichannel=False)  
radon_dictionary = iradon(radon(image))  

dictionaries = [wavelet_dictionary, radon_dictionary]
# -------------- Perform MCA Decomposition -------------- 
components = mca(image, dictionaries, alpha=0.1)  

# -------------- Visualization -------------- 
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray', vmin=0, vmax=1)
plt.title('Original Image')

for i, component in enumerate(components):
    plt.subplot(2, 3, i+2)
    plt.imshow(component, cmap='gray', vmin=0, vmax=1)
    plt.title(f'Component {i+1}')

plt.subplot(2, 3, 6) # Add a plot for the reconstruction
reconstructed = np.sum(components, axis=0)
plt.imshow(reconstructed, cmap='gray', vmin=0, vmax=1)
plt.title('Reconstructed Image')

plt.tight_layout()
plt.show()
```
