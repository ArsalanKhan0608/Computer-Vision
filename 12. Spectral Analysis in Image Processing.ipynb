```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Load the image
image_path = 'c130.jpg'
image = cv2.imread(image_path)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
def display_spectrum(image, title):
    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply FFT
    dft = cv2.dft(np.float32(gray_image), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    
    # Magnitude spectrum
    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))
    
    # Display the magnitude spectrum
    plt.figure(figsize=(6, 6))
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
    
    return dft_shift

# Display the Fourier Transform spectrum
dft_shifted = display_spectrum(image, 'Magnitude Spectrum')
```

```python
def band_pass_filter(dft_shift, low_freq, high_freq):
    # Create a mask with high value (1) at low frequency and high frequency regions
    rows, cols = dft_shift.shape[:2]
    crow, ccol = rows // 2, cols // 2
    mask = np.zeros((rows, cols, 2), np.uint8)
    mask[crow-low_freq:crow+low_freq, ccol-low_freq:ccol+low_freq] = 1
    mask[crow-high_freq:crow+high_freq, ccol-high_freq:ccol+high_freq] = 1
    
    # Apply mask and inverse DFT
    fshift = dft_shift * mask
    f_ishift = np.fft.ifftshift(fshift)
    img_back = cv2.idft(f_ishift)
    img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])
    
    # Display the filtered image
    plt.figure(figsize=(6, 6))
    plt.imshow(img_back, cmap='gray')
    plt.title('Band-pass Filtered Image')
    plt.axis('off')
    plt.show()
    
    return img_back

# Simulate spectral analysis by applying a band-pass filter
filtered_img = band_pass_filter(dft_shifted, low_freq=30, high_freq=60)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image_path = 'c130.jpg'
image = plt.imread(image_path)

# Display the original image
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
plt.show()

# Check if the image has an alpha channel and remove it
if image.shape[2] == 4:
    image = image[:, :, :3]

# Reshape the image to (number_of_pixels, number_of_bands)
num_pixels = image.shape[0] * image.shape[1]
num_bands = image.shape[2]
image_reshaped = image.reshape(num_pixels, num_bands)

# Apply PCA to reduce the number of bands from 3 to 2
pca = PCA(n_components=2)
image_pca = pca.fit_transform(image_reshaped)

# Reshape the transformed data back into a 2D image format
image_pca_reshaped = image_pca.reshape(image.shape[0], image.shape[1], 2)

# Display the image after PCA
plt.figure(figsize=(10, 6))
plt.imshow(image_pca_reshaped[:, :, 0], cmap='gray')
plt.axis('off')
plt.title('Image after PCA')
plt.show()

```

```python
from sklearn.cluster import KMeans
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'c130.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

# Preprocess the image for K-means clustering
pixels = image.reshape((-1, 3))

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(pixels)

# Replace each pixel value with its corresponding cluster centroid
segmented_img = kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape)
segmented_img = segmented_img.astype(np.uint8)

# Display the segmented image
plt.imshow(segmented_img)
plt.title('Segmented Image by K-means Clustering')
plt.axis('off')
plt.show()
```

```python
import pywt

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Perform a 2D Discrete Wavelet Transform
coeffs2 = pywt.dwt2(gray_image, 'haar')
cA, (cH, cV, cD) = coeffs2

# Show the results
fig = plt.figure(figsize=(12, 3))

for i, a in enumerate([cA, cH, cV, cD]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(['Approximation', 'Horizontal detail',
                  'Vertical detail', 'Diagonal detail'][i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
